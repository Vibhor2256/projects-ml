```python
import pandas as pd
```

```python
column_name= ['user_id', 'item_id', 'rating', 'timestamp']
path= 'C:/Users/Swarnima/Desktop/ML Minor Project/Movie Recommender/file.tsv'
```

```python
df= pd.read_csv(path, sep='\t', names=column_name)
df.head(6)
```

```python
movie_titles=pd.read_csv('C:/Users/Swarnima/Desktop/ML Minor Project/Movie Recommender/Movie_Id_Titles.csv')
movie_titles.head(6)
```

```python
data= pd.merge(df,movie_titles,on='item_id', how='left')
data.head()
```

```python
avg_rating=pd.DataFrame(data.groupby('title')['rating'].mean())
avg_rating.head(5)
```

```python
avg_rating['Total Rating']=pd.DataFrame(data.groupby('title')['rating'].count())
avg_rating.head()
```

```python
movie_user=data.pivot_table(index='user_id',columns='title',values='rating')
movie_user.head()
```

```python
correlations=movie_user.corrwith(movie_user['12 Angry Men (1957)'])
correlations.head(10)
```

```python
recommendations= pd.DataFrame(correlations,columns=['correlations'])
recommendations.dropna(inplace=True)
```

```python
recommendations1= pd.merge(recommendations,avg_rating['Total Rating'],on='title')
recommendations1.head()
```

```python
finalrec=recommendations1[recommendations1['Total Rating']>100].sort_values('correlations',ascending=False).reset_index()
```

```python
finalrec=finalrec.merge(movie_titles,on='title', how='right')
```
