```python
import numpy as np
import pandas as pd

df=pd.read_csv("C:\Study Material\Minor\STOCK-PREDICTION-USING-SVM-REGRESSION-master/AAPL.csv")
df.head()
```

```python
X=df.iloc[:,[1,2,3,6]]
X
```

```python
y=df.iloc[:,4]
y
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10 )

#train, validate, test = np.split(df.sample(frac=1), [int(.6*len(df)), int(.8*len(df))])
    

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,train_size=0.70)
#X_train, X_val, y_train, y_val = train_test_split(X_train,y_train,test_size = 0.25,train_size =0.75) 
```

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
X_train = sc1.fit_transform(X_train)
X_test = sc1.transform(X_test)
```

```python
from sklearn import svm
svm_model = svm.SVR(kernel='rbf', C=0.8)
svm_model.fit(X_train,y_train)
predictions_svm = svm_model.predict(X_test)
```

```python

u=svm_model.predict(X_test)
print('MAPE is :',np.mean(np.abs((y_test - u) / y_test)) * 100)
```

```python
y_test_pred=svm_model.predict(X_test)
from sklearn import metrics
mse_test = metrics.mean_squared_error(y_test,y_test_pred)
rmse_test= np.sqrt(mse_test)
print("Score (test MSE): {}".format(mse_test))
print("Score (test RMSE): {}".format(rmse_test))
```

```python
svm_model.score(X_test,y_test)
```
